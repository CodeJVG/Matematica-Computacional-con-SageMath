# Grafica de una función explicita

**Ejemplo:** Vamos a graficar la función $f(x) = x\cos\frac{1}{x}$

```python
f(x) = x*cos(1/x)
```

```python
f.plot()
```

```python
# Podemos asignar el dominio de la funcion
f.plot(-0.1,0.1)
```

```python
plot??
```

```python
plot(f, (x,-0.1,0.1), figsize=4)
```

```python
plot(f, (x,-0.1,0.1), figsize=4,color="green")
```

```python
plot(f, (x,-0.1,0.1), figsize=4,color="green",gridlines=True)
```

```python
plot(f, (x,-0.1,0.1), figsize=4,color="green",thickness=0.5,linestyle="--",title="Graph od xcos(1/x)")
```

## Graficando varias funciones juntas

**Ejemplo:** Graficar las funciones $y = x\cos\frac{1}{x}$,  $y=x$ y $y=-x$ juntos

```python
p = plot(f,(x,-0.1,0.1),figsize=6, color="green", legend_label="y=f(x)")
```

```python
p1 = plot(x,(x,-0.1,0.1),color="red", legend_label="y=x") 
```

```python
p2 = plot(-x,(x,-0.1,0.1),color="blue", legend_label="y=-x") 
```

```python
show(p+p1+p2,figsize=4)
```
